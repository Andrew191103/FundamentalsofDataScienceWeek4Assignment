```python
import pandas as pd
import numpy as np
from sklearn.impute import SimpleImputer
```

```python
df1 = pd.read_csv('/content/obese_nations_iso.csv')
```

```python
df2 = pd.read_csv('/content/obesity_nations_2014.csv')
```

```python
np.random.seed(50)
col_name=['value','Country']
mask = np.random.choice([True, False], size=df1[col_name].shape)
mask[mask.all(1),-1] = 0
```

```python
print("\nMask for df1:")
print(mask)
```

```python
df1[col_name]=df1[col_name].mask(mask)
```

```python
np.random.seed(10)
col_name=['Country']
mask = np.random.choice([True, False], size=df2[col_name].shape)
mask[mask.all(1),-1] = 0
print("\nMask for df2:")
print(mask)
```

```python
df2[col_name]=df2[col_name].mask(mask)
```

```python
print("\nMissing values in df1:")
print(df1.isnull().sum())
```

```python
print("\nMissing values in df2:")
print(df2.isnull().sum())
```

```python
df1['value'].fillna(df1['value'].mean(), inplace=True)
df1["Country"] = df1["Country"].fillna("America")
```

```python
df2["Country"] = df2["Country"].fillna("Australia")
```

```python
df1_melted = df1.melt(id_vars=["Country"], var_name="Variable", value_name="Value")
print("\nMelted df1:")
print(df1_melted)
```

```python
df_merged = df1.merge(df2, left_on='Country', right_on='Country')
print("\nMerged DataFrame:")
print(df_merged)
```

```python
df1.rename(columns={"value": "new_value"}, inplace=True)
```

```python
df1.set_index("Country", inplace=True)
```

```python
def custom_function(x):
    return x * 2
```

```python
df1['new_value'] = df1['new_value'].apply(custom_function)
```

```python
filtered_df = df1.query('new_value > 50')
print("\nFiltered DataFrame:")
print(filtered_df)
```

```python
df1.dropna(inplace=True)
```

```python
imputer = SimpleImputer(strategy='mean')
df1['new_value'] = imputer.fit_transform(df1[['new_value']])
```
